```python
import os
import time
import threading
from pymycobot.mycobot280 import MyCobot280
from pymycobot.genre import Angle, Coord
```

```python
mc = MyCobot280('/dev/ttyJETCOBOT', 1000000)
mc.thread_lock = True
print("로봇이 연결되었습니다.")
```

```python
initial_angles = [0,0,0,0,0,0]# 현재 좌표 확인
current_coords = mc.get_coords()
print("현재 좌표:", current_coords)
```

```python
initial_angles = [0,0,0,0,0,-50]
speed = 100

print("로봇을 초기 위치로 리셋합니다")
mc.send_angles(initial_angles, speed)

mc.set_gripper_value(100, 20)
time.sleep(2)
```

```python
joint_id = Angle.J1.value

angle = 40
speed = 10
print(f'{joint_id}를 {angle}로 이동합니다')

mc.send_angle(joint_id, angle, speed)
time.sleep(2)
```

```python
# Z축을 낮추기
work_coords = current_coords.copy()
work_coords[2] -= 10 # Z를 50mm 내리기
print(f"Z축을 {work_coords[2]}로 내립니다.")
mc.send_coords(work_coords, 30, 0)
time.sleep(2)
```

```python
# 제어할 joint선언
joint_id = Angle.J3.value
# 제어 각도와 속도
angle = -90
speed = 20
print(f'관절{joint_id}를 {angle}로 이동합니다')

mc.send_angle(joint_id, angle, speed)
time.sleep(2)
```
